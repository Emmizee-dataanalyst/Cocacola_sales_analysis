## Research Questions

1. Which beverage brand has the most sales?
2. Which beverage brand has the most profits?
3. Which beverage brand has the most units sold? 
4. Monthly trend of beverage brand by total sales? 
5. The state trend based on units sold. 
6. The beverage brand with the highest operating margin. 
7. Monthly trend by units sold. 

## Solutions

```python
# 1. Beverage brand with the most sales

beverage_sales = cola.groupby('Beverage Brand')['Total Sales'].sum().sort_values(ascending=False)
print(beverage_sales)
```

```python
# 2. Beverage brand with the most profits

beverage_profits = cola.groupby('Beverage Brand')['Operating Profit'].sum().sort_values(ascending=False)
print(beverage_profits)
```

```python
# 3. Beverage brand with the most units sold

beverage_sold = cola.groupby('Beverage Brand')['Units Sold'].sum().sort_values(ascending=False)
print(beverage_sold)
```

```python
# 4. Monthly trend of beverage brand by total sales

cola['Month'] = cola['Invoice Date'].dt.to_period('M').astype(str)

yearly_sales = cola.groupby(['Month', 'Beverage Brand'])['Total Sales'].sum().reset_index()
print(yearly_sales)
```

```python
# 5. The state trend based on units sold and total sales

state_summary = cola.groupby('State')[['Units Sold', 'Total Sales']].sum().sort_values(by='Total Sales', ascending=False)
print(state_summary)
```

```python
# 6. The beverage brand with the highest operating margin

beverage_margin = cola.groupby('Beverage Brand')['Operating Margin'].mean().sort_values(ascending=False)
print(beverage_margin)
```

```python
#7. Monthly trend by units sold

cola['Month'] = cola['Invoice Date'].dt.to_period('M').astype(str)

monthly_units = cola.groupby('Month')['Units Sold'].sum().reset_index().sort_values(by='Units Sold', ascending=False)
print(monthly_units)
```
